- Recency : 얼마나 최근에 구매했는가  
- Frequency : 얼마나 자주 구매했는가  
- Monetary : 얼마나 많은 금액을 지출했는가

```python
import pandas as pd
# 현재날짜를 가져오기 위해 datetime 모듈 import
import datetime

# 파일 경로 설정 및 데이터프레임화
path = './datasets/customer_shopping_data.csv'
customer_shopping_data_df = pd.read_csv(path)

display(customer_shopping_data_df)

# 1.Recency
## invoice_date컬럼의 값들을 날짜 형식으로 변환
customer_shopping_data_df['invoice_date'] = pd.to_datetime(customer_shopping_data_df['invoice_date'], dayfirst=True)

## invoice_date와 현재 날짜의 차이를 계산하기 위해 current_date를 선언하고, 기존의 data와 연산하기 위해 dataframe타입으로 변환 시켰다.
current_date = datetime.datetime.now().date()
current_date_df = pd.to_datetime(current_date, dayfirst=True)

## Recency 컬럼을 추가하여 현재 날짜와의 차이 계산
customer_shopping_data_df['Recency'] = (current_date_df - customer_shopping_data_df['invoice_date'])

# 2.Frequency
## 아래의 value_counts를 가져올때 Series타입으로 가져오기때문에, map으로 customer_id에 매핑하여 Frequency 컬럼을 생성
frequency_value = customer_shopping_data_df['customer_id'].value_counts()
customer_shopping_data_df['Frequency'] = customer_shopping_data_df['customer_id'].map(frequency_value)

# 3.Monetary
customer_shopping_data_df['Monetary'] = customer_shopping_data_df['price']

# 4.결과 확인
display(customer_shopping_data_df[['customer_id', 'Recency','Frequency','Monetary']])

# 5.나머지 컬럼 삭제
drop_columns = ['invoice_no','gender','age','category','quantity','price','payment_method','invoice_date','shopping_mall',]
customer_shopping_data_df = customer_shopping_data_df.drop(drop_columns, axis=1)

display(customer_shopping_data_df)
```
